```python
import numpy as np 
import os
import matplotlib.pyplot as plt
```

```python
# load frame 270 for example 
frame = 116

# load the data
person1 = np.load("xy_data/frame_{i}_1.npy".format(i=frame))
person2 = np.load("xy_data/frame_{i}_2.npy".format(i=frame))

# get the plot from detected_video/frame000270.jpg
img = plt.imread("detected_video/frame_00{frame}.jpg".format(frame=frame))
# show the plot with the person1 and person2's x and y coordinates as text. like person1 : (x,y), person2 : (x,y)

plt.text(1250,400, "person1 : ({:.2f}cm,{:.2f}cm)".format(person1[0],person1[1]), color="red", fontsize=10)
plt.text(750,300, "person2 : ({:.2f}cm,{:.2f}cm)".format(person2[0],person2[1]), color="red", fontsize=10)
plt.title("frame_{i}".format(i=frame))
plt.imshow(img)
```
